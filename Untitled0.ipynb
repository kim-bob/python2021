```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
ys= 200 + np.random.randn(100)
x= [x for x in range(len(ys))]

plt.plot(x, ys, '-') 
plt.fill_between(x, ys, 195, where=(ys>195), facecolor='r', alpha=0.6)

plt.title("sample ")
plt.show()
```
